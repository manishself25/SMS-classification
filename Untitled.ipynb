```python
### Data pre processing
```

- Lower Case
- Tokenization
- Removing Special Characters
- Removing stopwords and punctuation
- Stemming - Involve Lemmization

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import string
ps = PorterStemmer()
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word.tokenize(text)
    
    y = []
    for i in text:
        if i.isalnum():            # removing of special character
            y.append(i)
        
    text = y[:]                    # assign text to y
    y.clear()
    
    for i in text:
        if i not in stopwords.words("english") and i not in string.punctuation:
            y.append(i)
            
    text = y[:]                    # assign text to y
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
        
    return " ".join     # joining of list of data into a single stream
```

```python
transform_text("I loved the yt lecture on Machine Learning. How about You?")
```

```python
sfinal["transformed_text"] = sfinal["text"].apply(transform_text)
sfinal
```
